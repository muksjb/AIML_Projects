1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points)


```python
import pandas as pd

laliga = pd.read_csv("Laliga.csv",header=1)
```

3.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)


```python
def Goal_diff_count():
    GD = pd.DataFrame(pd.to_numeric(laliga["GoalsFor"]) - pd.to_numeric(laliga["GoalsAgainst"]), columns=['Goal Differences'])
    GD= GD.set_index(laliga.Team)
    return GD
```

```python
 Goal_diff_count()
```

Team with max Goal Difference

```python
DF =  Goal_diff_count()
```

```python
DF.loc[DF['Goal Differences']==DF['Goal Differences'].max(),:]
```

Team with Min Goal Difference

```python
DF.loc[DF['Goal Differences']==DF['Goal Differences'].min(),:]
```
